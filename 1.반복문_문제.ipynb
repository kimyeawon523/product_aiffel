### 실행 결과가 나오게끔 코드를 작성하세요.
- while / for 둘 다 작성하면서 연습을 해보세요!

1번 문제 : 1~30의 자연수 중 3의 배수만 출력해보세요.

for 작성

```python
for i in range(1, 31):
  if i % 3 == 0:
    print(i, end=' ')
```

while 작성

```python
num = 3
while num <=30:
  print(num, end=' ')
  num += 3
```

2번 문제 : 1~30의 자연수 중 3의 배수 중 6으로 나누어떨어지는 수는 제외하고 출력하세요.

for 작성

```python
for i in range(1, 31):
  if i % 6 != 0:
    print(i)
```

while 작성

```python
num = 1
while num <=30:
  if num % 6 != 0:
    print(num)
  num += 1
```

3번 문제 : 6명에 대한 python 점수가 리스트에 번호순으로 담겨있습니다. 번호 순서대로 **점수**와 **등급**을 출력하세요. (for문으로 작성)
- 80점 이상은 A, 60이상 80미만은 B, 60 미만은 C 등급입니다.

```python
python_score = [50, 85, 60, 93, 45, 79]

for i in python_score:
  if i >= 80:
    print("A입니다.")
  elif i >= 60:
    print("B입니다.")
  else:
    print("C입니다.")
```

4번 문제 : 사용자로부터 입력을 받아 -1을 입력받으면 입력을 종료하고 여태 입력한 사용자의 값을 전부 더해서 출력하세요. (while문으로 작성)
- 사용자가 -1을 바로 입력했을 경우 0을 출력합니다.

```python
"""

사용자 입력을 받는 방법
예시) int(input("값을 입력하세요."))

"""
total = 0

while True:
  user_input = int(input("값을 입력하세요"))

  if user_input == -1:
    break

  total += user_input

print(total)
```

```python
# 정답 코드
user_input = 0
total = 0
while user_input != -1:
  total += user_input
  user_input = int(input("값을 입력하세요"))

print(total)
```
