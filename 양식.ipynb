#### 1️⃣ 데이터과학자가 pandas를 배우는 이유

**핵심 포인트**: 빅데이터 시대의 데이터 전처리 필수 도구

**주요 특징:**

- 80~90%의 데이터 수집 및 정리 작업 담당
- 데이터과학의 가장 기초적이고 중요한 도구
- 클라우드 컴퓨팅 환경에서의 필수 프로그래밍 지식

#### 2️⃣ pandas 자료구조
📊 Series (1차원 배열)
- Series는 1차원 배열로 index와 value를 가진다.
- 딕셔너리를 Series로 바로 변환이 가능하며 Key는 index로 Value는 Value로 변환된다.
- class pandas.Series(data=None, index=None, dtype=None, name=None, copy=None, fastpath=<no_default>)

```python
# Series 생성
import pandas as pd

# 딕셔너리로 Series 생성
dict_data = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
sr = pd.Series(dict_data)
print(sr)

# 기본 구조 확인
print(f"인덱스: {sr.index}")
print(f"값: {sr.values}")
```

### 🔍 구체적인 차이점들

#### 1️⃣ 데이터 접근 방식

```python
# 리스트: 위치로만 접근 (의미 불분명)
seojun_scores_list = [90, 85, 95]
# 서준이 성적을 Series로
seojun_scores = pd.Series([90, 85, 95],
                         index=['수학', '국어', '영어'])

print(seojun_scores_list[0])# 90 (이게 뭔 과목?)
print(seojun_scores_list[1])# 85 (이것도 뭔 과목?)# Series: 의미있는 라벨로 접근
print(seojun_scores['수학'])# 90 (명확히 수학 점수!)
print(seojun_scores['국어'])# 85 (명확히 국어 점수!)
```

#### 2️⃣ 데이터 분석 편의성

```python
# 리스트: 수동으로 계산
total = sum(seojun_scores_list)
average = total / len(seojun_scores_list)
print(f"평균: {average}")

# Series: 내장 함수 활용
print(f"평균: {seojun_scores.mean()}")
print(f"최고점: {seojun_scores.max()}")
print(f"최고 과목: {seojun_scores.idxmax()}")# 최고점인 과목명까지!
```

#### 3️⃣ 조건 필터링

```python
# 리스트: 복잡한 과정
high_scores_list = []
for i, score in enumerate(seojun_scores_list):
    if score >= 90:
# 어떤 과목인지 별도로 관리해야 함
        subjects = ['수학', '국어', '영어']
        high_scores_list.append((subjects[i], score))

# Series: 간단하고 직관적
high_scores = seojun_scores[seojun_scores >= 90]
print("90점 이상 과목:")
print(high_scores)
```

### **🔥 미션 4-1 (종합 프로젝트)**
다음 온라인 쇼핑몰 판매 데이터를 분석해보세요:

**요구사항:**

1. 총 매출액을 계산하세요 (가격 × 수량)
2. 고객별 주문 횟수와 총 구매금액을 구하세요
3. 가장 많이 팔린 상품을 찾으세요
4. 일별 매출 현황을 분석하세요
5. 10만원 이상 구매한 주문만 필터링하세요

```python
import pandas as pd

# 샘플 데이터 생성
sales_data = {
    '주문번호': ['ORD001', 'ORD002', 'ORD003', 'ORD004', 'ORD005', 'ORD006'],
    '고객명': ['김철수', '이영희', '박민수', '김철수', '최지연', '이영희'],
    '상품명': ['노트북', '마우스', '키보드', '모니터', '노트북', '스피커'],
    '가격': [1200000, 25000, 80000, 300000, 1350000, 120000],
    '수량': [1, 2, 1, 1, 1, 1],
    '주문일': ['2024-01-15', '2024-01-16', '2024-01-16', '2024-01-17', '2024-01-17', '2024-01-18']
}

df_sales = pd.DataFrame(sales_data)
df_sales
```

```python
#1 총 매출액을 계산하세요 (가격 × 수량)

df_sales["가격"]*df_sales["수량"]
total = (df_sales["가격"]*df_sales["수량"]).sum()
total
```

```python
#2 고객별 주문횟수와 총 구매금액을 구하세요
# 고객별 주문횟수
# df_sales.groupby("고객명")["주문번호"].count()

df_sales["주문량"] = df_sales["가격"] * df_sales["수량"]


#agg를 사용하여 컬럼마다 다른 집계 적용할 수 있음
# 주문번호에는 count, 주문량 sum
result = df_sales.groupby("고객명").agg(
    order_count=("주문번호", "count"),
    total_purchase=("주문량", "sum"))


result
```

```python
#3 가장 많이 팔린 상품을 찾으세요

#답이 노트북, 마우스 2개가 나와야할 것 같은데 왜 노트북만 나올까...
#max()는 얼마나, idxmax()는 누가
df_sales.groupby("상품명")["수량"].sum().idxmax()
```

```python
#4 일별 매출 현황을 분석하세요

# to_datetime : 문자열이나 숫자로 된 날짜데이터를
# datetime 타입으로 바꿔서 날짜 연산을 가능하게 해줌
#df_sales["주문일"] = pd.to_datetime(df_sales["주문일"])

#주문별 매출
df_sales["판매"] = df_sales["가격"] * df_sales["수량"]

#일자별 집계
daily = df_sales.groupby("주문일").agg(
    order_count=("주문번호", "count"),          # 주문 건수
    total_quantity=("수량", "sum"),            # 판매 수량 합
    total_sales=("판매", "sum"),               # 총 매출
    unique_customers=("고객명", "nunique"),    # 구매 고객 수( nunique 중복 제거)
).reset_index()

print(daily)
```

```python
#5 10만원 이상 구매한 주문만 필터링하세요

df_sales[df_sales["가격"]*df_sales["수량"] > 100000]
```

### **심화 미션1: 결측치 처리**
다음 데이터에서 결측치를 처리해보세요

**요구사항:**

1. 결측치가 있는 행과 열을 확인하세요
2. 나이의 결측치는 평균값으로 채우세요
3. 점수의 결측치는 중앙값으로 채우세요

```python
import numpy as np

# 결측치가 있는 데이터
# 결측치란 관측되거나 입력되어야 할 데이터가 비어 있거나 존재하지 않는 값
# 대표적인 형태 NaN(Not a Number) / None / NaT 날짜, 시간 결측치

messy_data = {
    '이름': ['A', 'B', 'C', 'D', 'E'],
    '나이': [25, np.nan, 30, 28, np.nan],
    '점수': [85, 90, np.nan, 95, 88]
}

df_messy = pd.DataFrame(messy_data)
df_messy
```

```python
#1 결측치가 있는 행과 열을 확인하세요
# isnull : 결측치 True, 정상값 False

df_messy.isnull()
```

```python
#2 나이의 결측치는 평균값으로 채우세요
#평균을 계산 후 반올림하여 평균값으로 채움

df_messy["나이"]=df_messy["나이"].fillna(df_messy["나이"].mean().round())
display(df_messy)

```

```python
#3 점수의 결측치는 중앙값으로 채우세요


df_messy["점수"] = df_messy["점수"].fillna(df_messy["점수"].median())
display(df_messy)
```

```python
# 기본정보
students = pd.DataFrame({
    '학번': ['S001', 'S002', 'S003'],
    '이름': ['김철수', '이영희', '박민수'],
    '학과': ['컴퓨터공학', '통계학', '경영학']
})

# 성적정보
scores = pd.DataFrame({
    '학번': ['S001', 'S002', 'S003'],
    '수학': [85, 92, 78],
    '영어': [88, 85, 90]
})

# 병합
merged = pd.merge(students, scores, on='학번')

# 평균 점수 계산
merged['평균'] = merged[['수학', '영어']].mean(axis=1)
print(merged)
```
