#### 1-2. Series 이해하기

```python
# 학생들의 점수 데이터로 Series 만들기
data = {
    '김철수': 85,
    '이영희': 92,
    '박민수': 78,
    '최지연': 96
}

score_series = pd.Series(data)

print(score_series)
```

```python
# 이영희의 점수 출력하기
print(score_series['이영희'])

# 90점 이상인 학생들을 필터링하기
print(score_series[score_series >= 90])

# 모든 점수에 5점씩 더하기
score_series = score_series + 5
print(score_series)
```

### 3. 데이터 조작과 필터링

#### 3-1. 인덱싱과 슬라이싱

```python
# 1. loc를 사용해서 이영희의 모든 정보를 출력하세요
print(df.loc[1])

# 2. iloc를 사용해서 첫 번째와 세 번째 행의 '이름'과 '점수'를 출력하세요
print(df.iloc[0, 2], [0, 3])

# 3. 나이가 20살인 학생들을 필터링하세요
age = df[df['나이'] == 20]

print(age)
```

#### 3-2. 데이터 수정과 추가

```python
# 1. 새로운 학생 '정민호(22세, 남, 88점)'를 추가하세요
df_work = df.copy()
df_work.loc[4] = ['정민호', 22, '남', 88]
print(df_work)

# 2. 모든 학생의 점수에 3점을 더하세요
df['점수'] = df['점수'] + 3

print(df)

# 3. '등급' 열을 추가하세요 (90점 이상 : A, 80-89점 : B, 80점 미만 : C)
df.loc[df['점수'] >= 90, '등급'] = 'A'
df.loc[(df['점수'] >= 80) & (df['점수'] < 90), '등급'] = 'B'
df.loc[df['점수'] < 80, '등급'] = 'C'

print(df_work)
```

- df.copy() : 복사본 만들기 (dnjsqhs epdlxj qhgh)
- loc[행, 열] : 이름(라벨)을 기준으로 특정 데이터 선택
- iloc[행, 열] : 순서를 기준으로 특정 데이터 선택
- and, or 대신 &, | 사용

### 4. 실전 데이터 분석

```python

import pandas as pd

# 샘플 데이터 생성
sales_data = {
    '주문번호': ['ORD001', 'ORD002', 'ORD003', 'ORD004', 'ORD005', 'ORD006'],
    '고객명': ['김철수', '이영희', '박민수', '김철수', '최지연', '이영희'],
    '상품명': ['노트북', '마우스', '키보드', '모니터', '노트북', '스피커'],
    '가격': [1200000, 25000, 80000, 300000, 1350000, 120000],
    '수량': [1, 2, 1, 1, 1, 1],
    '주문일': ['2024-01-15', '2024-01-16', '2024-01-16', '2024-01-17', '2024-01-17', '2024-01-18']
}

df_sales = pd.DataFrame(sales_data)
```

```python
# 1. 총 매출액 계산 (가격 x 수량)
df_sales['매출액'] = df_sales['가격'] * df_sales['수량']

print(df_sales['매출액'])
```

```python
# 2. 고객별 주문 횟수와 총 구매금액
result = df_sales.groupby('고객명').agg({'주문번호': 'count', '매출액': 'sum'})

print(result)
```

```python
# 3. 가장 많이 팔린 상품
most_sold_item = df_sales.groupby('상품명')['수량'].sum().idxmax()

print(most_sold_item)
```

```python
# 4. 일별 매출 현황 분석
daily_sales = df_sales.groupby('주문일')['매출액'].sum()

print(daily_sales)
```

```python
# 5. 10만원 이상 구매한 주문만 필터링
filtered_orders = df_sales[df_sales['가격'] * df_sales['수량'] >= 100000]

print(filtered_orders)
```

- groupby() : 특정 열을 기준으로 데이터 묶기
- agg() : 그룹화된 데이터에 여러 연산 동시 적용
- mean() : 평균
- sum() : 합계
- idmax() : 가장 큰 값을 가진 행의 이름을 반환
- axis = 1 : 가로
- axis = 0 : 세로

### 추가 심화 미션

#### 심화 미션 1. 결측치 처리

```python
import numpy as np

# 결측치가 있는 데이터
messy_data = {
    '이름': ['A', 'B', 'C', 'D', 'E'],
    '나이': [25, np.nan, 30, 28, np.nan],
    '점수': [85, 90, np.nan, 95, 88]
}

df_messy = pd.DataFrame(messy_data)
```

```python
# 1. 결측치가 있는 행과 열을 확인
print(df_messy.isnull().sum())

print(df_messy[df_messy.isnull().any(axis=1)])
```

```python
# 2. 나이의 결측치는 평균값으로 채우기
df_messy['나이'] = df_messy['나이'].fillna(df_messy['나이'].mean())

print(df_messy)
```

```python
# 3. 점수의 결측치는 중앙값으로 채우기
df_messy['점수'] = df_messy['점수'].fillna(df_messy['점수'].median())

print(df_messy)
```

- isnull() : 비어있는 값인지 확인
- fillna() : 결측치를 특정 값으로 채우기

#### 심화 미션 2. 데이터 병합

```python
# 학생 기본정보
students = pd.DataFrame({
    '학번': ['S001', 'S002', 'S003'],
    '이름': ['김철수', '이영희', '박민수'],
    '학과': ['컴퓨터공학', '통계학', '경영학']
})

# 학생 성적정보
scores = pd.DataFrame({
    '학번': ['S001', 'S002', 'S003'],
    '수학': [85, 92, 78],
    '영어': [88, 85, 90]
})
```

```python
# 1. 두 DataFrame을 학번 기준으로 병합
df_merged = pd.merge(students, scores, on='학번')

print(df_merged)
```

```python
# 2. 각 학생의 평균 점수를 계산하세요
df_merged['평균'] = df_merged[['수학', '영어']].mean(axis=1)

print(df_merged)
```

- df_merge : 두 개의 표를 공통 열(key) 기준으로 하나로 합치기
