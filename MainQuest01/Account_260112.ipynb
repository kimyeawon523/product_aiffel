```python
# 1. Account 클래스

import random

class Account:
  account_count = 0

  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance
    self.bank_name = ""
    self.account_number = ""
    
    self.deposit_count = 0  # 7. 이자 지급 변수   
    self.deposit_log = []
    self.withdraw_log = []

    self.open_account() # 2. 계좌 개수 변수
    Account.account_count += 1

# 2.클래스 변수
  def open_account(self):
    self.bank_name = "SC은행"
    num1 = random.randint(100, 999)
    num2 = random.randint(10,99)
    num3 = random.randint(100000, 999999)
    self.account_number = f"{num1}-{num2}-{num3}"

    print(f"\n---계좌가 개설되었습니다.---")
    print(f"은행 이름 : {self.bank_name}")
    print(f"예금주  : {self.owner}")
    print(f"계좌번호: {self.account_number}")
    print(f"잔액    : {self.balance}원")


# 3. 클래스 변수 출력
  @classmethod
  def get_account_num(cls):
    return cls.account_count


# 4. 입금 메서드
  def deposit(self, amount):
    if amount >= 1:
      self.balance += amount
      self.deposit_count += 1
      self.deposit_log.append(amount)
      print(f"{amount}원이 입금되었습니다. \n현재 잔액 : {self.balance}원")
# 7. 이자 지급하기
      if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
                print(f"5회 입금 기념 이자가 입금되었습니다.")
                print(f"이자 금액 {format(interest, ',')}원 / 현재 잔액 {format(int(self.balance), ',')}원")
      else:
          print(f"현재 잔액 : {format(int(self.balance), ',')}원")
    else:
      print("입금 실패! 1원 이상만 입금할 수 있습니다.")

  def info(self):
    print(f"\n---계좌 정보---")
    print(f"예금주 : {self.owner} / 잔액 : {self.balance}원")


# 5. 출금 메서드
  def withdraw(self, amount):
    if self.balance >= amount:
      if amount >= 1:
        self.balance -= amount
        self.withdraw_log.append(amount)
        print(f"{amount}원이 출금되었습니다. \n현재 잔액 : {self.balance}원")
      else:
        print("출금 실패! 1원 이상만 출금할 수 있습니다.")

    else:
      print(f"출금 실패! 잔액이 부족합니다. \n현재 잔액 : {self.balance}원")

# 6. 정보 출력 메서드
  def display_info(self):
    formatted_balance = format(self.balance, ",")  # {변수명:,}을 사용하면 숫자에 자동으로 3자리마다 쉼표(,)가 들어간다.


# 10. 입금과 출금 내역이 기록
    # 입금 내역
  def deposit_history(self):
          print(f"\n--- [{self.owner}님 입금 내역] ---")
          if not self.deposit_log:
            print("입금 내역이 없습니다.")
          else:
            for i, amount in enumerate(self.deposit_log, 1):
              print(f"{i}회차: {amount:,}원")

    # 출금 내역
  def withdraw_history(self):
      print(f"\n--- [{self.owner}님 출금 내역] ---")
      if not self.withdraw_log:
          print("출금 내역이 없습니다.")
      for i, amount in enumerate(self.withdraw_log, 1):
          print(f"{i}회차: {amount:,}원")

# 8. 여러 객체 생성
account_list = [
    Account("김예원", 1500000), 
    Account("안선미", 50000),  
    Account("김경국", 2500000),
    Account("유지숙", 300000)]

# 9. 특정 고객의 정보 출력
for acc in account_list:
  if acc.balance >= 1000000:
    acc.display_info()
    print("\n--- 잔고 100만원 이상 고객 명단---")
    print(f"{acc.owner}님 계좌 잔액 : {acc.balance:,}원 입니다")

print(f"\n 현재까지 생성된 총 계좌 수: {Account.get_account_num()}개")
print(f"현재 리스트에 저장된 고객 수: {len(account_list)}명")

            
```
